```python
import pandas as pd
import pickle as pkl
```

This is a simple workbook used to generate the starting point for the weighting process. The tables in the workbook were in a strange format so I just did it manually here.

```python
all_rows_post = pd.read_pickle('all_rows_post2.pkl')
```

```python
all_rows_post = all_rows_post.groupby(['yearSale','Sector','Technology','HeatingEfficiency','CoolingEfficiency']).sum().reset_index()
all_rows_post = all_rows_post[all_rows_post['Sector'] == 'Residential']
key_techs = ['Heat Pump','Heat Pump - Ductless','Central Air Conditioning - Condenser','Gas Furnace']
all_rows_post = all_rows_post[all_rows_post['Technology'].isin(key_techs)]
all_rows_post = all_rows_post.drop(['Reported Quantity','Supplier','extrapped_qty','TOTAL'], axis=1)
all_rows_post = all_rows_post.replace('NA','UNDEFINED')
```

```python
all_rows_post.to_pickle('sales_mix.pkl')
```
