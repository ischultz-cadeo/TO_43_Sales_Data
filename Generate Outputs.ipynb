```python
import pandas as pd
import pickle as pkl
```

This workbook generates output CSV tables for the Executive summary, memo & final workbook results. Starting point is the 'all_rows_post' pickled dataframe generated in the main analysis notebook. 

```python
all_rows_post = pd.read_pickle('all_rows_post.pkl')
```

```python
detailed_sales = pd.read_excel('Extrap_byCZ_HARDI_Adjusted_25JAN2022_IsaacCopy.xlsx' , sheet_name = 'DetailedSales', usecols='A:H')
reported_sales = pd.read_excel('Extrap_byCZ_HARDI_Adjusted_25JAN2022_IsaacCopy.xlsx' , sheet_name = 'memo - reported', usecols='N:R')
```

```python
detailed_sales =detailed_sales[detailed_sales['Supplier'] != 0]
```

```python
all_rows_post.to_csv("All Rows Post Extrapolations.csv")
```

```python
extrapolated_quantities=all_rows_post.groupby(['yearSale','Technology'])['TOTAL'].sum().reset_index()
#extrapolated_quantities2=all_rows_post.groupby(['yearSale','Technology','Sector'])['TOTAL'].sum().reset_index()
reported_quantities=detailed_sales.groupby(['yearSale','sector','Technology'])['Reported Quantity'].sum().reset_index()
```

```python
#extrapolated_quantities2.to_csv("EXtrapolated_Quantities with sector.csv")
extrapolated_quantities.to_csv("EXtrapolated_Quantities.csv")
reported_quantities.to_csv("Reported_Quantities.csv")
```

```python
ductless_heat_pump=all_rows_post[all_rows_post['Technology'] == 'Heat Pump - Ductless']
ductless_heat_pump = ductless_heat_pump[ductless_heat_pump['Sector'] == 'Residential']
heating = ductless_heat_pump.groupby(['yearSale','HeatingEfficiency'])['TOTAL'].sum().reset_index()
cooling = ductless_heat_pump.groupby(['yearSale','CoolingEfficiency'])['TOTAL'].sum().reset_index()


years = [2016, 2017, 2018, 2019, 2020]
heating_eff_mix = pd.DataFrame()
cooling_eff_mix = pd.DataFrame()


for y in years:
    df1 = heating[heating['yearSale'] == y]
    df2 = cooling[cooling['yearSale'] == y]
    total1=df1['TOTAL'].sum()
    total2=df2['TOTAL'].sum()
    df1['pct_share'] = df1['TOTAL'] / total1
    df2['pct_share'] = df2['TOTAL'] / total2
    heating_eff_mix = pd.concat([heating_eff_mix, df1], axis=0)
    cooling_eff_mix = pd.concat([cooling_eff_mix, df2], axis=0)
    
heating_eff_mix.to_csv("Ductless Heat Pump Heating Eff Mix.csv")
cooling_eff_mix.to_csv("Ductless Heat Pump Cooling Eff Mix.csv")
```

```python
cac = all_rows_post[all_rows_post['Technology'] == 'Central Air Conditioning - Condenser']
mix = cac.groupby(['yearSale','CoolingEfficiency'])['TOTAL'].sum().reset_index()


years = [2016, 2017, 2018, 2019, 2020]
cooling_mix = pd.DataFrame()

for y in years:
    df = mix[mix['yearSale'] == y]
    total=df['TOTAL'].sum()
    df['pct_share'] = df['TOTAL'] / total
    cooling_mix = pd.concat([cooling_mix, df], axis=0)
    
cooling_mix.to_csv("CAC- Condenser Cooling Eff Mix.csv")
```

```python
gas_furnace = all_rows_post[all_rows_post['Technology'] == 'Gas Furnace']
gas = gas_furnace.groupby(['yearSale','HeatingEfficiency'])['TOTAL'].sum().reset_index()


years = [2016, 2017, 2018, 2019, 2020]
heating_mix = pd.DataFrame()

for y in years:
    df = gas[gas['yearSale'] == y]
    total=df['TOTAL'].sum()
    df['pct_share'] = df['TOTAL'] / total
    heating_mix = pd.concat([heating_mix, df], axis=0)
    
heating_mix.to_csv("Gas Furnace Eff Mix.csv")
```

```python
extrapolation_summary1=all_rows_post.groupby(['yearSale','Technology'])['TOTAL'].sum().reset_index()
extrapolation_summary2=all_rows_post.groupby(['yearSale','Technology'])['extrapped_qty'].sum().reset_index()
extrapolation_summary = extrapolation_summary1.merge(extrapolation_summary2, how='outer')
extrapolation_summary['pct_extrapolated'] = extrapolation_summary['extrapped_qty'] / extrapolation_summary['TOTAL']
extrapolation_summary.to_csv("Pct Extrapolated Summary.csv")
```

```python
extrapolation_summary3=all_rows_post.groupby(['yearSale'])['TOTAL'].sum().reset_index()
extrapolation_summary4=all_rows_post.groupby(['yearSale'])['extrapped_qty'].sum().reset_index()
extrapolation_summary5 = extrapolation_summary3.merge(extrapolation_summary4, how='outer')
extrapolation_summary5['pct_extrapolated'] = extrapolation_summary5['extrapped_qty'] / extrapolation_summary5['TOTAL']
extrapolation_summary5.to_csv("Pct Extrapolated Summary by Year.csv")
```

```python
res  = all_rows_post2[all_rows_post2['Sector'] == 'Residential']
res1 = res[(res['Technology'] == 'Heat Pump - Ductless') | (res['Technology'] == 'Heat Pump')|(res['Technology'] == 'Gas Furnace')]
res2 = res[(res['Technology'] == 'Heat Pump') | (res['Technology'] == 'Central Air Conditioning - Condenser')]
res1=res1[res1['HeatingEfficiency'] != 'UNDEFINED']
res_heating = res1.groupby(['yearSale','Technology','HeatingEfficiency'])['TOTAL'].sum().reset_index()
res_cooling = res2.groupby(['yearSale','Technology','CoolingEfficiency'])['TOTAL'].sum().reset_index()


years = [2016, 2017, 2018, 2019, 2020]
res_heat_eff_mix = pd.DataFrame()
res_cool_eff_mix = pd.DataFrame()


for y in years:
    df1 = res_heating[res_heating['yearSale'] == y]
    df2 = res_cooling[res_cooling['yearSale'] == y]
   
    ductless_total  = df1[df1['Technology'] == 'Heat Pump - Ductless']['TOTAL'].sum()
    hp_total        = df1[df1['Technology'] == 'Heat Pump']['TOTAL'].sum()
    furnace_total   = df1[df1['Technology'] == 'Gas Furnace']['TOTAL'].sum()
    cac_total       = df2[df2['Technology'] == 'Central Air Conditioning - Condenser']['TOTAL'].sum()

    df1.loc[df1['Technology'] == 'Heat Pump', 'pct_share']            = df1['TOTAL'] / hp_total
    df1.loc[df1['Technology'] == 'Heat Pump - Ductless', 'pct_share'] = df1['TOTAL'] / ductless_total
    df1.loc[df1['Technology'] == 'Gas Furnace', 'pct_share']          = df1['TOTAL'] / furnace_total
    
    df2.loc[df2['Technology'] == 'Heat Pump', 'pct_share'] = df2['TOTAL'] / hp_total
    df2.loc[df2['Technology'] == 'Central Air Conditioning - Condenser', 'pct_share'] = df2['TOTAL'] / cac_total
    
    res_heat_eff_mix = pd.concat([res_heat_eff_mix, df1], axis=0)
    res_cool_eff_mix = pd.concat([res_cool_eff_mix, df2], axis=0)
    
res_heat_eff_mix.to_csv("Residential Heating Eff Mix.csv")
res_cool_eff_mix.to_csv("Residential Cooling Eff Mix.csv")
```
